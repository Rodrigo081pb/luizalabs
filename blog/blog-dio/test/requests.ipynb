```python
import requests

url_api_post = "http://127.0.0.1:8000/posts"

new_post = {
    "title": "Meu novo post via API",
    "date": "2024-06-01T12:00:00Z",
    "published": True
}

response = requests.post(url_api_post, json=new_post)

print("Status Code:", response.status_code)
print("Response Body:", response.json())
```

```python
import requests

url = "http://127.0.0.1:8000/posts?published=true&limit=10"

response = requests.get(url)

print(response.status_code)
print(response.json())
```
